## Visualization
---

```python
font_path = '/Library/Fonts/AppleGothic.ttf'

def show_wordcloud(df: pd.DataFrame):
    texts = df.title
    plt.figure(figsize = (10,10))
    wc = WordCloud(width = 1600, height = 800,
                collocations=False, font_path=font_path).generate(" ".join(texts))
    plt.imshow(wc)
```

```python
worst = df[df.rating == 1]
not_worst = df[df.rating > 1]

show_wordcloud(worst)
```

```python
show_wordcloud(not_worst)
```
