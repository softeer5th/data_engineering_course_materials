### SQL SELECT DISTINCT Statement


- 중복 제거에는 다양한 방법이 있음.
1. GROUP BY
대상: 집계가 필요한 테이블 (판매, 로그, 통계 데이터)
상황: COUNT, SUM, AVG 등의 집계가 필요할 때

```sql
-- 예: 국가별 총 주문액
SELECT Country, COUNT(*) as OrderCount, SUM(Amount) as TotalAmount
FROM Orders
GROUP BY Country
```

2. ROW_NUMBER() OVER(PARTITION BY)
대상: 타임스탬프가 있는 테이블 (이벤트 로그, 센서 데이터)
상황: 시간순으로 가장 최근 데이터만 필요할 때

```sql
-- 예: 각 고객의 가장 최근 주문
SELECT * FROM (
    SELECT *, ROW_NUMBER() OVER (PARTITION BY CustomerID ORDER BY OrderDate DESC) as rn
    FROM Orders
) tmp
WHERE rn = 1
```

3. EXISTS
대상: 대용량 테이블
상황: 인덱스가 잘 구성되어 있고, 전체 칼럼이 필요할 때 (왜지?)

```sql
-- 예: 중복된 이메일을 가진 첫 번째 사용자만 선택
SELECT a.* 
FROM Users a
WHERE NOT EXISTS (
    SELECT 1 FROM Users b
    WHERE b.Email = a.Email 
    AND b.ID < a.ID
)
```

4. UNION
대상: 여러 테이블의 데이터를 합칠 때
상황: 테이블 구조는 같지만 데이터가 분산되어 있을 때

```sql
-- 예: 현재 고객과 이전 고객 데이터 통합
SELECT CustomerID, Name, Email FROM CurrentCustomers
UNION
SELECT CustomerID, Name, Email FROM ArchivedCustomers
```

5. DISTINCT
대상: 작은 테이블
상황: 단순한 중복 제거가 필요하고 성능이 중요하지 않을 때

```sql
-- 예: 주문이 있는 고객 목록
SELECT DISTINCT CustomerID 
FROM Orders
```

````python
with sqlite3.connect('northwind.db') as conn:
    # Select DISTINCT
    print("SELECT DISTINCT")
    cursor.execute(
        '''
        SELECT DISTINCT Country FROM Customers;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # SELECT Example Without DISTINCT
    print("SELECT Example Without DISTINCT")
    cursor.execute(
        '''
        SELECT Country FROM Customers;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    
    # COUNT DISTINCT
    print("COUNT DISTINCT")
    cursor.execute(
        '''
        SELECT COUNT(DISTINCT Country) FROM Customers;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # IN MS Access
    print("IN MS Access")
    cursor.execute(
        '''
        SELECT COUNT(*) AS DistinctCountries
        FROM (SELECT DISTINCT Country FROM Customers);
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    conn.commit()
````

- DISTINCT는 SELECT 오른쪽의 모든 값이 동일할 때 중복으로 판단함.

### SQL WHERE Clause
WHERE는 records를 필터링하는 데 사용됨.

- SELECT 뿐 아니라 UPDATE, DELETE 문에서도 사용 가능.

````python
with sqlite3.connect('northwind.db') as conn:
    # The SQL WHERE Clause

    print("The SQL WHERE Clause")
    cursor.execute(
        '''
        SELECT *  FROM Customers
        WHERE Country='Mexico';
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # Text Fields vs. Numeric Fields

    print("Text Fields vs. Numeric Fields")
    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerID=1;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # Operators in The WHERE Clause
    print("Operators > in The WHERE Clause")
    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerID > 4;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    print("Operators > in The WHERE Clause")
    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerID > 4;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    print("Operators <> in The WHERE Clause") # Not Equal
    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerID <> 4;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    print("Operators BETWEEN in The WHERE Clause")
    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerName BETWEEN "A" AND "C";
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    print("Operators LIKE in The WHERE Clause") 
    # Indexing을 위하여 앞에 %를 붙이는 것은 좋지 않다. 뒤에 붙이는 것이 좋다.
    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerName LIKE "A%";
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
        
    print("Operators IN in The WHERE Clause")
    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerName IN ("Alfreds Futterkiste", "Ana Trujillo Emparedados y helados");
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL ORDER BY Keyword
ORDER BY는 결과를 정렬하는 데 사용됨.

```sql
SELECT column1, column2, ...
FROM table_name
ORDER BY column1, column2, ... ASC|DESC;
```

````python
with sqlite3.connect('northwind.db') as conn:
    # The SQL ORDER BY

    print("The SQL ORDER BY")
    cursor.execute(
        '''
        SELECT * FROM Products
        ORDER BY Price;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # DESC

    print("DESC")
    cursor.execute(
        '''
        SELECT * FROM Products
        ORDER BY Price DESC;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Order Alphabetically

    print("Order Alphabetically")
    cursor.execute(
        '''
        SELECT * FROM Products
        ORDER BY ProductName;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Alphabetically DESC

    print("Alphabetically DESC") 
    cursor.execute(
        '''
        SELECT * FROM Products
        ORDER BY ProductName DESC; 
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # ORDER BY Several Columns

    print("ORDER BY Several Columns") 
    
    cursor.execute(
        '''
        SELECT * FROM Customers
        ORDER BY Country, CustomerName; 
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Using Both ASC and DESC

    print("Using Both ASC and DESC") 

    cursor.execute(
        '''
        SELECT * FROM Customers
        ORDER BY Country ASC, CustomerName DESC; 
        '''
    )
    
    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL AND Operator
AND는 WHERE 절에 두 개 이상의 조건을 지정하는 데 사용됨.

```sql
SELECT column1, column2, ...
FROM table_name
WHERE condition1 AND condition2 AND condition3 ...;
```

OR은 하나라도 참이면 됨.

````python
with sqlite3.connect('northwind.db') as conn:
    cursor = conn.cursor()

    # The SQL AND Operator

    print("The SQL AND Operator") 
    
    cursor.execute(
        '''
        SELECT * 
        FROM Customers
        WHERE Country = 'Spain' AND CustomerName LIKE 'G%';
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # All Conditions Must Be True

    print("All Conditions Must Be True") 

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE Country = 'Germany'
        AND City = 'Berlin'
        AND PostalCode > 12000;
        '''
    )
    
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Combining AND and OR

    print("Combining AND and OR") 

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE Country = 'Spain' AND (CustomerName LIKE 'G%' OR CustomerName LIKE 'R%');
        '''
    )
    
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Without Parentheses

    print("Without Parentheses") 
    # AND 절과 OR 절이 별개가 된다.
    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE Country = 'Spain' AND CustomerName LIKE 'G%' OR CustomerName LIKE 'R%';
        '''
    )
    
    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL NULL Values

- NULL 값은 필드에 데이터가 없음을 의미함. 0이나 공백과 다름.

````python
with sqlite3.connect('northwind.db') as conn:
    cursor = conn.cursor()

    # The IS NULL Operator

    print("The IS NULL Operator") 
    
    cursor.execute(
        '''
        SELECT CustomerName, ContactName, Address
        FROM Customers
        WHERE Address IS NULL;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # The IS NOT NULL Operator

    print("The IS NOT NULL Operator") 

    cursor.execute(
        '''
        SELECT CustomerName, ContactName, Address
        FROM Customers
        WHERE Address IS NOT NULL;
        '''
    )
    
    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL UPDATE Statement
- 데이터베이스의 기존 레코드를 업데이트하는 데 사용됨.

```sql
UPDATE table_name
SET column1 = value1, column2 = value2, ...
WHERE condition;
```

- [주의] WHERE 절을 생략하면 모든 레코드가 업데이트됨.

````python
with sqlite3.connect('northwind.db') as conn:
    cursor = conn.cursor()

    # The SQL UPDATE Statement

    print("The SQL UPDATE Statement") 
    
    cursor.execute(
        '''
        UPDATE Customers
        SET ContactName = 'Alfred Schmidt', City= 'Frankfurt'
        WHERE CustomerID = 1;
        '''
    )
    affected_rows = cursor.rowcount
    print(f"영향받은 행의 수: {affected_rows}")

    conn.rollback()
    
    # Update Multiple Columns

    print("Update Multiple Columns") 

    cursor.execute(
        '''
        UPDATE Customers
        SET ContactName='Juan'
        WHERE Country='Mexico';
        '''
    )
    affected_rows = cursor.rowcount
    print(f"영향받은 행의 수: {affected_rows}")

    conn.rollback()
    
    # Update Warning!

    print("Update Warning!") 

    cursor.execute(
        '''
        UPDATE Customers
        SET ContactName='Juan';
        '''
    )
    affected_rows = cursor.rowcount
    print(f"영향받은 행의 수: {affected_rows}")

    conn.rollback()
````

### SQL TOP, LIMIT, FETCH FIRST or ROWNUM Clause

- TOP은 상위 n개의 레코드를 선택하는 데 사용됨. 큰 테이블에서 확인할 때 유용함. 
- MySQL에선 LIMIT

```sql
SELECT TOP number|percent column_name(s)
FROM table_name
WHERE condition;
```

Oracle 12
```sql
SELECT column_name(s)
FROM table_name
ORDER BY column_name(s)
FETCH FIRST number ROWS ONLY;
```

````python
with sqlite3.connect('northwind.db') as conn:
    cursor = conn.cursor()

    # LIMIT

    print("LIMIT") 
    
    cursor.execute(
        '''
        SELECT * FROM Customers
        LIMIT 3;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # FETCH FIRST (Oracle)

    print("FETCH FIRST (Oracle)") 

    # cursor.execute(
    #     '''
    #     SELECT * FROM Customers
    #     FETCH FIRST 3 ROWS ONLY;
    #     '''
    # )
    
    # rows = cursor.fetchall()
    # for row in rows:
    #     print(row)
    
    # SQL TOP PERCENT Example

    print("SQL TOP PERCENT Example (SQL Server/MS Access)") 

    # cursor.execute(
    #     '''
    #     SELECT TOP 50 PERCENT * FROM Customers;
    #     '''
    # )
    
    # rows = cursor.fetchall()
    # for row in rows:
    #     print(row)

    # Oracle Percent

    """
    SELECT * FROM Customers
    FETCH FIRST 50 PERCENT ROWS ONLY;
    """

    # ADD a WHERE CLAUSE

    print("ADD a WHERE CLAUSE (SQL Server, MS Access)") 
    
    '''
    SELECT TOP 3 * FROM Customers
    WHERE Country='Germany';
    '''

    print("ADD a WHERE CLAUSE (MySQL)") 

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE Country='Germany'
        LIMIT 3;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    print("ADD a WHERE CLAUSE (Oracle)") 
    
    '''
    SELECT * FROM Customers
    WHERE Country='Germany'
    FETCH FIRST 3 ROWS ONLY;
    '''

    # ADD the ORDER BY Keyword

    print("ADD the ORDER BY Keyword (MySQL)") 

    cursor.execute(
        '''
        SELECT * FROM Customers
        ORDER BY CustomerName DESC
        LIMIT 3;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL MIN() and MAX() Functions


````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # MIN Example
    print("MIN Example")

    cursor.execute(
        '''
        SELECT MIN(Price)
        FROM Products;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # MAX Example
    print("MAX Example")

    cursor.execute(
        '''
        SELECT MAX(Price)
        FROM Products;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # Set Column Name (Alias)
    print("Set Column Name (Alias)")

    cursor.execute(
        '''
        SELECT MIN(Price) AS SmallestPrice
        FROM Products;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Use MIN() with GROUP BY
    print("Use MIN() with GROUP BY")
    print("CategoryID 기준 Grouping 후 가장 작은 Price")

    cursor.execute(
        '''
        SELECT MIN(Price) AS SmallestPrice, CategoryID
        FROM Products
        GROUP BY CategoryID;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### The SQL COUNT() Function
```sql
SELECT COUNT(*)
FROM Products;
```

- Column을 지정하면 NULL이 아닌 값만 Count한다.

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # Specify Column
    print("Specify Column")

    cursor.execute(
        '''
        SELECT COUNT(ProductName)
        FROM Products;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Add a WHERE Clause
    print("Add a WHERE Clause")

    cursor.execute(
        '''
        SELECT COUNT(ProductID)
        FROM Products
        WHERE Price > 20;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # Ignore Duplicates
    print("Ignore Duplicates")

    cursor.execute(
        '''
        SELECT COUNT(DISTINCT Price)
        FROM Products;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Use an Alias
    print("Use an Alias")
    
    cursor.execute(
        '''
        SELECT COUNT(*) AS [Number of records]
        FROM Products;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    
    # Use COUNT() with GROUP BY
    print("Use COUNT() with GROUP BY")
    
    cursor.execute(
        '''
        SELECT COUNT(*) AS [Number of records], CategoryID
        FROM Products
        GROUP BY CategoryID;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL LIKE Operator
- WHERE 절에서 패턴 매칭을 위해 사용됨.
- %는 0개 이상의 문자와 매칭됨.
- _는 한 문자와 매칭됨.

```sql
SELECT * FROM Customers
WHERE CustomerName LIKE 'a%';
```

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # The SQL LIKE Operator
    print("The SQL LIKE Operator")

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerName LIKE 'a%';
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # The _ Wildcard
    print("The _ Wildcard")

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE city LIKE 'L_nd__';
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
   
    # The % Wildcard
    print("The % Wildcard")

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE city LIKE '%L%';
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
   
    # Starts With
    print("Starts With")

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerName LIKE 'La%';
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Ends With
    print("Ends With")

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerName LIKE '%a';
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Contains
    print("Contains")

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerName LIKE '%or%';
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
   
    # Combine Wildcards
    print("Combine Wildcards")

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE CustomerName LIKE 'a__%';
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Without Wildcard
    print("Without Wildcard: db 설정에 따라 대소문자 구분하지 않음. = 과 같음")

    cursor.execute(
        '''
        SELECT * FROM Customers
        WHERE Country LIKE 'Spain';
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL Wildcard Characters

### SQL Aliases
- table, column에 대한 별칭을 지정함.
- AS는 생략 가능함.

사용이 용이한 경우:
- 컬럼명이 길거나 복잡한 경우
- 여러 테이블을 조인할 때
- 집계 함수를 사용할 때
- 서브쿼리를 사용할 때
- 두 개 이상의 컬럼을 합칠 때

```sql
SELECT CustomerID AS ID
FROM Customers;
```

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # SQL Aliases
    print("SQL Aliases")

    cursor.execute(
        '''
        SELECT CustomerID AS ID
        FROM Customers;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # AS is Optional
    print("AS is Optional")

    cursor.execute(
        '''
        SELECT CustomerID ID
        FROM Customers;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
   
    # Alias for Columns
    print("Alias for Columns")

    cursor.execute(
        '''
        SELECT CustomerID AS ID, CustomerName AS Customer
        FROM Customers;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Using Aliases With a Space Character

    print("Using Aliases With a Space Character")

    cursor.execute(
        '''
        SELECT ProductName AS [My Great Products]
        FROM Products;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Double Quotes도 가능.
    print("Double Quotes로는 띄어쓰기 포함 가능.")
    cursor.execute(
        '''
        SELECT ProductName AS "My Great Products"
        FROM Products;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)


    # Concatenate Columns

    print("Concatenate Columns")

    cursor.execute(
        '''
        SELECT CustomerName, Address + ', ' + PostalCode + ' ' + City + ', ' + Country AS Address
        FROM Customers;
        '''
    )
    

    # Concatenate Columns (MySQL)

    print("Concatenate Columns (MySQL). sqlite는 안됨.")

    '''
    SELECT CustomerName, CONCAT(Address,', ',PostalCode,', ',City,', ',Country) AS Address
    FROM Customers;
    '''

    # Oracle Example
    '''
    SELECT CustomerName, (Address || ', ' || PostalCode || ' ' || City || ', ' || Country) AS Address
    FROM Customers;
    '''

    #  Alias for Tables

    print("Alias for Tables")

    cursor.execute(
        '''
       SELECT * FROM Customers AS Persons;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    #  More shorter queries with Alias

    print("More shorter queries with Alias")

    cursor.execute(
        '''
        SELECT o.OrderID, o.OrderDate, c.CustomerName
        FROM Customers AS c, Orders AS o
        WHERE c.CustomerName='Around the Horn' AND c.CustomerID=o.CustomerID;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    #  Same but without Alias

    print("Same but without Alias")

    cursor.execute(
        '''
        SELECT Orders.OrderID, Orders.OrderDate, Customers.CustomerName
        FROM Customers, Orders
        WHERE Customers.CustomerName='Around the Horn' AND Customers.CustomerID=Orders.CustomerID;
        '''
    )

    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL JOIN

- 두 개 이상의 테이블을 결합하는 데 사용됨.
- 여러 형태의 JOIN이 있음.
- INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL JOIN, CROSS JOIN
- (INNER) JOIN은 두 테이블 간의 key가 일치하는 레코드를 반환함.
- LEFT (OUTER) JOIN은 왼쪽 테이블의 모든 레코드와 오른쪽 테이블의 일치하는 레코드를 반환함.
- RIGHT (OUTER) JOIN은 오른쪽 테이블의 모든 레코드와 왼쪽 테이블의 일치하는 레코드를 반환함.

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # SQL JOIN
    print("SQL JOIN")

    cursor.execute(
        '''
        SELECT Orders.OrderID, Customers.CustomerName, Orders.OrderDate
        FROM Orders
        INNER JOIN Customers ON Orders.CustomerID=Customers.CustomerID;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
````

### SQL INNER JOIN

- 컬럼에 테이블 이름을 지정해주는 것은 좋은 관행이다
- INNER JOIN과 JOIN은 동일하다.
```sql
SELECT column_name(s)
FROM table1
INNER JOIN table2
ON table1.column_name = table2.column_name
```

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # SQL JOIN
    print("SQL JOIN")

    cursor.execute(
        '''
        SELECT ProductID, ProductName, Products.CategoryID
        FROM Products
        INNER JOIN Categories ON Products.CategoryID = Categories.CategoryID;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # Naming the Columns
    print("Naming the Columns: 컬럼에 테이블 이름을 지정해주는 것은 좋은 관행이다.")

    cursor.execute(
        '''
        SELECT Products.ProductID, Products.ProductName, Categories.CategoryName
        FROM Products
        INNER JOIN Categories ON Products.CategoryID = Categories.CategoryID;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # JOIN or INNER JOIN
    print("JOIN or INNER JOIN")

    cursor.execute(
        '''
        SELECT Products.ProductID, Products.ProductName, Categories.CategoryName
        FROM Products
        JOIN Categories ON Products.CategoryID = Categories.CategoryID;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # JOIN Three Tables
    print("JOIN Three Tables")

    cursor.execute(
        '''
        SELECT Orders.OrderID, Customers.CustomerName, Shippers.ShipperName
        FROM ((Orders
        INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID)
        INNER JOIN Shippers ON Orders.ShipperID = Shippers.ShipperID);
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
````

### SQL UNION ALL Operator
- UNION과 동일하지만 중복된 레코드도 모두 반환함.

```sql
SELECT column_name(s) FROM table1
UNION ALL
SELECT column_name(s) FROM table2;
```

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # SQL UNION Example
    print("SQL UNION Example: 중복이 없다")

    cursor.execute(
        '''
        SELECT City FROM Customers
        UNION
        SELECT City FROM Suppliers
        ORDER BY City;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # SQL UNION ALL Example
    print("SQL UNION ALL Example: 중복이 있다.")

    cursor.execute(
        '''
        SELECT City FROM Customers
        UNION ALL
        SELECT City FROM Suppliers
        ORDER BY City;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # SQL UNION With WHERE
    print("SQL UNION With WHERE: 중복이 있다.")

    cursor.execute(
        '''
        SELECT City, Country FROM Customers
        WHERE Country='Germany'
        UNION
        SELECT City, Country FROM Suppliers
        WHERE Country='Germany'
        ORDER BY City;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
   
    # SQL UNION ALL With WHERE
    print("SQL UNION ALL With WHERE: 중복이 있다.")

    cursor.execute(
        '''
        SELECT City, Country FROM Customers
        WHERE Country='Germany'
        UNION ALL
        SELECT City, Country FROM Suppliers
        WHERE Country='Germany'
        ORDER BY City;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # Another UNION Example
    print("Another UNION Example: Alias를 사용하여 컬럼명을 바꾸고, 어느 소속인지 알 수 있게 한다.")

    cursor.execute(
        '''
        SELECT 'Customer' AS Type, ContactName, City, Country
        FROM Customers
        UNION
        SELECT 'Supplier', ContactName, City, Country
        FROM Suppliers;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
````

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # SQL GROUP BY Examples
    print("SQL GROUP BY Examples")

    cursor.execute(
        '''
        SELECT COUNT(CustomerID), Country
        FROM Customers
        GROUP BY Country;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # 나라 별 많은 수 순서로 정렬
    print("나라 별 많은 수 순서로 정렬")

    cursor.execute(
        '''
        SELECT COUNT(CustomerID), Country
        FROM Customers
        GROUP BY Country
        ORDER BY COUNT(CustomerID) DESC;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # GROUP BY With JOIN Example
    print("GROUP BY With JOIN Example")

    cursor.execute(
        '''
        SELECT s.ShipperName, COUNT(o.OrderID) AS NumberOfOrders FROM Orders o
        LEFT JOIN Shippers s ON o.ShipperID = s.ShipperID
        GROUP BY ShipperName;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
````

### SQL HAVING Clause
- Having은 GROUP BY 절과 함께 사용됨.
- WHERE는 레코드를 필터링하는 데 사용되고, HAVING은 그룹을 필터링하는 데 사용됨.

```sql
SELECT column_name(s)
FROM table_name
WHERE condition
GROUP BY column_name(s)
HAVING condition
ORDER BY column_name(s);
```

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # SQL HAVING Examples
    print("SQL HAVING Examples")

    cursor.execute(
        '''
        SELECT COUNT(CustomerID), Country
        FROM Customers
        GROUP BY Country
        HAVING COUNT(CustomerID) > 5;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
 
    cursor.execute(
        '''
        SELECT COUNT(CustomerID), Country
        FROM Customers
        GROUP BY Country
        HAVING COUNT(CustomerID) > 5
        ORDER BY COUNT(CustomerID) DESC;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
    
    # More HAVING Examples
    print("More HAVING Examples")

    cursor.execute(
        '''
        SELECT Employees.LastName, COUNT(Orders.OrderID) AS NumberOfOrders
        FROM (Orders
        INNER JOIN Employees ON Orders.EmployeeID = Employees.EmployeeID)
        GROUP BY LastName
        HAVING COUNT(Orders.OrderID) > 10;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)

    # WHERE 절 추가
    print("WHERE 절 추가")

    cursor.execute(
        '''
        SELECT Employees.LastName, COUNT(Orders.OrderID) AS NumberOfOrders
        FROM Orders
        INNER JOIN Employees ON Orders.EmployeeID = Employees.EmployeeID
        WHERE LastName = 'Davolio' OR LastName = 'Fuller'
        GROUP BY LastName
        HAVING COUNT(Orders.OrderID) > 25;
        '''
    )
    rows = cursor.fetchall()
    print(f"조회된 행의 수: {len(rows)}")
    for row in rows:
        print(row)
````

### SQL ANY and ALL Operators
- 서브쿼리의 결과가 ANY 또는 ALL 조건과 일치하는지 확인하는 데 사용됨.

  (SELECT column_name

  WHERE condition);

### SQL NULL Functions
- NULL 값을 처리하는 데 사용됨.
- ISNULL(), COALESCE(), IFNULL(), NVL()
- MySQL은 IFNULL(), Oracle은 NVL(), SQL Server는 ISNULL()을 사용함.
- COALESCE()는 모든 DBMS에서 사용 가능함.

```sql
SELECT ProductName, UnitPrice * (UnitsInStock + COALESCE(UnitsOnOrder, 0))
FROM Products;
```

````python
with sqlite3.connect("demo.db") as conn:
    cursor = conn.cursor()

    # 기존 테이블이 있다면 삭제
    cursor.execute('DROP TABLE IF EXISTS Products')

    # 테이블 생성
    create_table_sql = '''
    CREATE TABLE Products (
        P_Id INTEGER PRIMARY KEY,
        ProductName TEXT NOT NULL,
        UnitPrice DECIMAL(10,2),
        UnitsInStock INTEGER,
        UnitsOnOrder INTEGER
    )
    '''
    cursor.execute(create_table_sql)

    # 데이터 삽입
    products_data = [
        (1, 'Jarlsberg', 10.45, 16, 15),
        (2, 'Mascarpone', 32.56, 23, None),
        (3, 'Gorgonzola', 15.67, 9, 20)
    ]

    cursor.executemany('INSERT INTO Products VALUES (?,?,?,?,?)', products_data)

    # 변경사항 저장
    conn.commit()

    # SQL NULL 미처리 문제
    print("SQL NULL 미처리 문제: NULL이 포함되어 있으면 결과가 NULL이 된다.")

    cursor.execute(
        '''
        SELECT ProductName, UnitPrice * (UnitsInStock + UnitsOnOrder)
        FROM Products;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # Solutions
    print("Solutions: COALESCE")

    cursor.execute(
        '''
        SELECT ProductName, UnitPrice * (UnitsInStock + COALESCE(UnitsOnOrder, 0))
        FROM Products;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
````

### SQL Stored Procedures for SQL Server
- sqlite3는 지원하지 않음. 대신 user defined function을 사용함.
- 저장 프로시저는 데이터베이스에 저장된 SQL 코드 블록임.
- 재사용 가능하고, 성능이 좋음.
- 매개변수를 사용하여 저장 프로시저를 호출할 수 있음.
- 저장 프로시저는 CREATE PROCEDURE 문을 사용하여 생성함.

```sql
CREATE PROCEDURE procedure_name
AS
sql_statement
GO;
```

- 저장 프로시저는 EXECUTE 문을 사용하여 호출함.

```sql
EXEC procedure_name;
```

````python
with sqlite3.connect("demo.db") as conn:
    cursor = conn.cursor()

    # Stored Procedure Example
    print("Stored Procedure Example")

    '''
    CREATE PROCEDURE SelectAllCustomers
    AS
    SELECT * FROM Customers
    GO;
    '''
    
    '''
    EXEC SelectAllCustomers;
    '''
    
    

    # Stored Procedure With One Parameter
    print("Stored Procedure With One Parameter")

    '''
    CREATE PROCEDURE SelectAllCustomers @City nvarchar(30)
    AS
    SELECT * FROM Customers WHERE City = @City
    GO;
    '''
    

    '''
    EXEC SelectAllCustomers @City = 'London';
    '''
   
    # Stored Procedure With Multiple Parameters
    print("Stored Procedure With Multiple Parameters")

    '''
    CREATE PROCEDURE SelectAllCustomers @City nvarchar(30), @PostalCode nvarchar(10)
    AS
    SELECT * FROM Customers WHERE City = @City AND PostalCode = @PostalCode
    GO;
    '''
    
    '''
    EXEC SelectAllCustomers @City = 'London', @PostalCode = 'WA1 1DP';
    '''
````

### 주석 달기

- 한 줄 주석은 --로 시작함.
- 여러 줄 주석은 /*로 시작하여 */로 끝남.

````python
with sqlite3.connect("northwind.db") as conn:
    cursor = conn.cursor()
    # Single Line Comments
    print("Single Line Comments")

    cursor.execute(
        '''
        -- Select all:
        SELECT * FROM Customers;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    cursor.execute(
        '''
        SELECT * FROM Customers -- WHERE City='Berlin';
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    
    # Multi-line Comments
    print("Multi-line Comments")

    cursor.execute(
        '''
        SELECT * FROM Customers WHERE (CustomerName LIKE 'L%'
        OR CustomerName LIKE 'R%' /*OR CustomerName LIKE 'S%'
        OR CustomerName LIKE 'T%'*/ OR CustomerName LIKE 'W%')
        AND Country='USA'
        ORDER BY CustomerName;
        '''
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
````
