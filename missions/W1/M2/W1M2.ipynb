```python
def sendQuery(sql):
    try:
        with sqlite3.connect("Northwind.db") as conn:
            cursor.execute(sql)
            conn.commit() #변경사항 저장
            return cursor.fetchall()[:10] #0~9번째 행만 읽어 반환
        
    except sqlite3.Error as e:
        print(e)
```

```python
#SQL Syntax
#SQL문은 간단한 키워드들로 구성됨
sql = "SELECT * FROM Customers;"
sendQuery(sql)
```

```python
#SQL SELECT
#데이터베이스에서 특정 데이터를 선택
# Customers 테이블에서 고객명과 도시를 셀렉
sql = "SELECT CustomerName, City FROM Customers;"
sendQuery(sql)
```

```python
#SQL SELECT DISTICT 
#고유한 값들만 반환
# Customers 테이블의 나라 항목을 중복을 없애 반환
sql = "SELECT DISTINCT Country FROM Customers;"
sendQuery(sql)
```

```python
#SQL Where
# 지정한 조건을 충족하는 레코드를 필터링하여 반환
# Customers 테이블의 레코드 중 나라가 멕시코에 해당하는 것만 반환
sql = """SELECT * FROM Customers
WHERE Country='Mexico';"""
sendQuery(sql)
```

```python
#SQL Order By
# 레코드 집합을 정렬하는데 사용
# Products의 레코드를 가격 기준 오름차순 정렬하여 반환
sql = """SELECT * FROM Products
ORDER BY Price;"""
sendQuery(sql)
```

```python
#SQL And
# WHERE절 응용 - 여러개의 조건
# Customers 테이블에서 나라가 스페인이면서 손님이름이 G로 시작하는 레코드 반환
sql = """SELECT *
FROM Customers
WHERE Country = 'Spain' AND CustomerName LIKE 'G%';"""
sendQuery(sql)
```

```python
#SQL Or
# WHERE절 응용 - 여러개의 조건
# Customers 테이블에서 나라가 독일 또는 스페인인 레코드만 반환
sql = """SELECT *
FROM Customers
WHERE Country = 'Germany' OR Country = 'Spain';"""
sendQuery(sql)
```

```python
#SQL Not
# 다른 연산자와 함께 사용되어 반대 결과 제공
# Customers 테이블에서 나라가 스페인이 아닌 레코드만 반환
sql = """SELECT * FROM Customers
WHERE NOT Country = 'Spain';"""
sendQuery(sql)
```

```python
#SQL Insert Into
# 테이블에 새로운 레코드를 삽입
# 고객명, 접속명, 주소, 도시, 우편번호, 나라 열에 새로운 레코드를 추가
sql = """INSERT INTO Customers (CustomerName, ContactName, Address, City, PostalCode, Country)
VALUES ('Cardinal', 'Tom B. Erichsen', 'Skagen 21', 'Stavanger', '4006', 'Norway');"""
sendQuery(sql)
```

```python
#SQL Null Values
# 값이 없는 필드는 NULL값을 갖는다
# Customers 테이블에서 주소값이 비어있지 않은 레코드들의 고객명, 접속명, 주소 반환
sql = """SELECT CustomerName, ContactName, Address
FROM Customers
WHERE Address IS NOT NULL;"""
sendQuery(sql)
```

```python
#SQL Update
# 원래 있던 레코드 수정
# Customers 테이블의 고객아이디가 1인 레코드를 새 접속명과 새 도시로 변경
sql = """UPDATE Customers
SET ContactName = 'Alfred Schmidt', City= 'Frankfurt'
WHERE CustomerID = 1;"""
sendQuery(sql)
```

```python
#SQL Select Top
# 반환할 레코드의 수를 지정
# sql = """SELECT TOP 3 * FROM Customers;""" 은 안됨. 맞는 구문이 있음!!!
# Customers의 3개 레코드를 반환
sql = """SELECT * FROM Customers LIMIT 3"""
sendQuery(sql)
```

```python
#SQL Aggregate Functions
# 집계함수는 어떤 값 집합에 대한 계산을 수행, 단일값을 반환
# Customers 테이블의 우편번호 행 수를 반환
sql = """SELECT COUNT(PostalCode) FROM Customers"""
sendQuery(sql)
```

```python
#SQL Min and Max
# 선택 열의 가장 작은/ 가장 큰 값을 반환
# Products 테이블의 가격 열 중 가장 작은 값을 반환
sql = """SELECT MIN(Price)
FROM Products;"""
sendQuery(sql)
```

```python
#SQL Count
# 지정한 기준에 맞는 행의 수 반환
# Products 테이블의 모든 레코드 행 수 반환
sql = """SELECT COUNT(*)
FROM Products;"""
sendQuery(sql)
```

```python
#SQL Sum
# 숫자형 열의 총 합 반환
# OrderDetails 테이블의 Quantity열의 총 합 반환
sql = """SELECT SUM(Quantity)
FROM OrderDetails;"""
sendQuery(sql)
```

```python
#SQL AVG
# 숫자형 열의 평균값 반환
# Products 테이블의 가격 열 평균값 반환
sql = """SELECT AVG(Price)
FROM Products;"""
sendQuery(sql)
```

```python
#SQL LIKE
# 와일드카드와 함께 사용해 특정 패턴을 찾음
# Customers에서 고객명이 a로 시작하는 레코드 반환
sql = """SELECT * FROM Customers
WHERE CustomerName LIKE 'a%';"""
sendQuery(sql)
```

```python
#SQL Wildcards
# 문자열엥서 하나 이상의 문자를 대체하는데 사용
# %, _, [], ^, -, {} 등
# Customers 테이블에서 고객명이 a로 시작하는 레코드 반환
sql = """SELECT * FROM Customers
WHERE CustomerName LIKE 'a%';"""
sendQuery(sql)
```

```python
#SQL In
# WHERE절에 여러개의 값을 지정 가능
# Customers 테이블에서 나라가 독일, 프랑스, 영국인 모든 레코드 반환
sql = """SELECT * FROM Customers
WHERE Country IN ('Germany', 'France', 'UK');"""
sendQuery(sql)
```

```python
#SQL BETWEEN
# 주어진 범위 내에서 값을 선택, 숫자말고도 텍스트 또는 날짜일 수 있다!!
# 시작과 끝 값은 포함됨
# Products 테이블에서 가격이 10이상 20이하인 레코드 반환
sql = """SELECT * FROM Products
WHERE Price BETWEEN 10 AND 20;"""
sendQuery(sql)
```

```python
#SQL Aliases - 별명
# 테이블이나 열에 별명을 저장하여 더 알기 쉽게 만듦.
# 쿼리가 지속되는 동안만 존재
# Customers 테이블의 고객아이디를 ID라는 별칭으로 셀렉
sql = """SELECT CustomerID AS ID
FROM Customers;"""
sendQuery(sql)
```

```python
#SQL Joins
# 관련 열을 기준으로 행을 결합함
# Customers 테이블과 Orders 테이블의 고객 아이디가 일치하면 주문번호, 고객명, 주문일자가 결합된 행들을 생성
sql = """SELECT Orders.OrderID, Customers.CustomerName, Orders.OrderDate
FROM Orders
INNER JOIN Customers ON Orders.CustomerID=Customers.CustomerID;"""
sendQuery(sql)
```

```python
#SQL Inner Join
# 두 테이블에서 일치하는 값을 갖는 레코드 선택
# Products 테이블과 Categories 테이블에서 카테고리명이 일치하면 제품아이디, 제품명, 카테고리명이 결합된 행들을 생성
sql = """SELECT ProductID, ProductName, CategoryName
FROM Products
INNER JOIN Categories ON Products.CategoryID = Categories.CategoryID;"""
sendQuery(sql)
```

```python
#SQL Left Join
# 왼쪽 테이블의 모든 레코드와, 오른쪽 테이블의 일치 레코드를 반환
# Customers의 모든 레코드와 고객아이디가 같은 Order의 레코드 중 고객명과 주문아이디를 반환
sql = """SELECT Customers.CustomerName, Orders.OrderID
FROM Customers
LEFT JOIN Orders ON Customers.CustomerID = Orders.CustomerID
ORDER BY Customers.CustomerName;"""
sendQuery(sql)
```

```python
#SQL Right Join
# 오른쪽 테이블의 모든 레코드와, 왼쪽 테이블의 일치 레코드를 반환
# Employees의 모든 레코드와 고용자아이디가 같은 Orders의 레코드 중 주문아이디, 고용자 라스트네임과 퍼스트네임 반환
sql = """SELECT Orders.OrderID, Employees.LastName, Employees.FirstName
FROM Orders
RIGHT JOIN Employees ON Orders.EmployeeID = Employees.EmployeeID
ORDER BY Orders.OrderID;"""
sendQuery(sql)
```

```python
#SQL Full Join
# 왼쪽 또는 오른쪽 테이블에 일치 항목이 있는 경우 모든 레코드 반환
# Union과 같은 기능을 함
# Customer과 Orders의 고객아이디를 기준으로 모든 값을 합집합 후 고객명, 주문아이디 반환- 고객명을 기준으로 오름차순 정렬
sql = """SELECT Customers.CustomerName, Orders.OrderID
FROM Customers
FULL OUTER JOIN Orders ON Customers.CustomerID=Orders.CustomerID
ORDER BY Customers.CustomerName;"""
# sql = """Select * From Customers where not ifnull(CustomerID, 1) and ifnull(CustomerName, 1)"""
sendQuery(sql)
```

```python
#SQL Self Join
# 테이블이 자체적으로 조인
# Customers 테이블에서 고객명은 다르지만 같은 도시에 사는 두 사람의 고객명1, 고객명2를 반환 - 고객1의 도시로 오름차순 정렬
sql = """SELECT A.CustomerName AS CustomerName1, B.CustomerName AS CustomerName2, A.City
FROM Customers A, Customers B
WHERE A.CustomerID <> B.CustomerID
AND A.City = B.City
ORDER BY A.City;"""
sendQuery(sql)
```

```python
#SQL Union
# 두 개 이상 명령문의 결과 집합을 결합
# 두 명령의 열은 같은 수, 비슷한 유형, 같은 순서여야함.
# Customers의 도시와 Suppliers의 도시를 가져와서 합집합 - 도시 기준 오름차순 정렬
sql = """SELECT City FROM Customers
UNION
SELECT City FROM Suppliers
ORDER BY City;"""
sendQuery(sql)
```

```python
#SQL Group By
# 동일한 값을 갖는 행을 그룹화
# Customers를 나라를 기준으로 그룹화, 각 나라의 고객아이디 수와 나라 반환
sql = """SELECT COUNT(CustomerID), Country
FROM Customers
GROUP BY Country;"""
sendQuery(sql)
```

```python
#SQL Having
# WHERE은 집계 함수와 사용할 수 없으므로 HAVNIG을 사용
# Customers를 나라를 기준으로 그룹화, 각 나라의 고객아이디 수와 나라 반환 - 고객아이디수가 5 초과인 것들만
sql = """SELECT COUNT(CustomerID), Country
FROM Customers
GROUP BY Country
HAVING COUNT(CustomerID) > 5;"""
sendQuery(sql)
```

```python
#SQL Exists
# 하위 쿼리에 레코드가 존재하는지 확인
# 하위 쿼리: Products 테이블에서 Products와 Suppliers의 공급자아이디가 같으면서 가격이 20미만인 제품명 반환
# 상위 쿼리: 제품이 있으면 공급자명 반환
# 즉, 물건을 공급한 적 있으면서 제품 가격이 20 미만인 공급원의 공급자명을 반환
sql = """SELECT SupplierName
FROM Suppliers
WHERE EXISTS (SELECT 1 FROM Products WHERE Products.SupplierID = 
Suppliers.supplierID AND Price < 20);"""
sendQuery(sql)
```

```python
#SQL Any, All
# Any - bool값 반환, 하나라도 조건 충족시 True
# All - bool값 반환, 모든 값이 조건 충족시 True
# 하위 쿼리: OrderDetails에서 수량이 10인 제품의 제품아이디들을 반환
# 상위 쿼리: Products에서 하위쿼리에서 반환한 제품아이디와 일치하는(where절이 true가 됨) 제품 반환

# SQLite에는 ANY연산자가 없다!!! 대신 IN 연산자를 이용할 수 있음
# ProductID = ANY(...) 대신 ProductID IN (...)로
sql = """SELECT ProductName
FROM Products
WHERE ProductID IN
(SELECT ProductID
FROM OrderDetails
WHERE Quantity = 10);"""
sendQuery(sql)
```

```python
# SQLite는 All도 지원하지 않음
sql = """SELECT ProductName
FROM Products
WHERE ProductID = ALL
(SELECT ProductID
FROM OrderDetails
WHERE Quantity = 10);"""
sendQuery(sql)
```

```python
#SQL Select Into
# 특정 열을 새 테이블에 복사
# Customers의 고객명, 접속명 열을 새 테이블 CustomersBackup2017에 복사

# SQLite는 Select Into를 지원하지 않는다!!! 대신 그냥 새로운 테이블을 Create
sql = """CREATE TABLE CustomersBackup2017 AS SELECT CustomerName, ContactName
FROM Customers;"""
sendQuery(sql)
```

```python
#SQL Insert Into Select
# 한 테이블의 데이터를 복사하여 다른 테이블에 삽입 - 대상 테이블의 기존 레코드는 영향을 받지 않음
# 소스 테이블과 대상 테이블의 데이터 유형이 일치해야 함 
# Customers의 고객명, 도시, 나라 열에 Suppliers의 공급자명, 도시, 나라를 복사하여 붙임)
sql = """INSERT INTO Customers (CustomerName, City, Country)
SELECT SupplierName, City, Country FROM Suppliers;"""
sql = """"""
sendQuery(sql)
```

```python
#SQL Case
# 조건이 맞으면 값을 반환
# 수량이 30이상이면 첫번째 문구, 30이면 두번째 문구, 30미만이면 3번째 문구를 반환
# END AS는 반환받은 문구들에 대해 원하는 컬럼명을 지정함 (AS 생략해도 같은 기능)
sql = """SELECT OrderID, Quantity,
CASE
    WHEN Quantity > 30 THEN 'The quantity is greater than 30'
    WHEN Quantity = 30 THEN 'The quantity is 30'
    ELSE 'The quantity is under 30'
END AS QuantityText
FROM OrderDetails;"""
sendQuery(sql)
```

```python
#SQL Null Functions
# IFNULL()이나 ISNULL()은 식이 null인 경우 대체값을 반환할 수 있다.
# Products에서 제품명과 price * unit을 셀렉. 이때 unit이 null값이어도 결과가 null이 되지 않도록 0으로 대체
sql = """SELECT ProductName, Price * IFNULL(Unit, 0)
FROM Products;"""
sendQuery(sql)
```
