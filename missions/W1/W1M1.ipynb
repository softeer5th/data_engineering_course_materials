### 1. 데이터셋을 분석해서 얻을 수 있는 경제적 가치는 무엇일까? 어떤 비즈니스 상황에서 이런 데이터셋을 분석해서 어떤 경제적 가치를 얻을 수 있을까?

- 상관관계는 원인과 결과는 아니다.-> 기법적인 측면에서 경제적이다. (원인에 따른 결과를 직접 도출하기에는 비용과 시간이 많이 듦)

- 데이터 값에 따른 상관관계 분석을 통해 고객에게 제품 이해도를 향상시키는 데 도움을 줄 수 있을 것이다.

- 데이터셋으로 기초통계량(평균값 등) 알 수 있으므로 최적의 값이 있다면 이 값으로 품질에 대한 측정이 가능할 것 

- 데이터셋의 비례, 반비례 관계들을 통해 자동차의 특성에 대해 수치적으로/직관적으로 알게 되었다. 비례/반비례 관계 에서 대부분이 추세를 따라가지만 튀는 값이 있을 경우, 특수한 니즈에 맞는 차량 판매가 가능

- 기업마다 비례/반비례의 계수 값이 다르면 그것에 대한 점을 찾아볼 수 있을 것 

### 2. 변수들 간의 상관 관계가 높은 조합을 임의로 2개 선택해서 해당 데이터 간의 상관 관계를 그래프로 그리고 어떤 결론을 내릴 수 있는지를 토의하세요.

Mpg: 연비
cyl: 실린더 개수
Disp: 엔진 실린더 부피(배기량)
Wt: 무게
Hp:엔진 최대 출력량
vs: 엔진의 구조

연비는 높을수록 좋다. 연비 높으면 실린더 개수가 적다. 실린더 개수 적으면 배기량도  작다. 무게도 작다 </br>
-> 작은차 - 연비 좋다 (실제로도 Wt - mpg(-0.86))

Mpg , cyl(-0.85) : 연비와 실린더 개수 반비례 관계?
Disp, cyl(0.90): 배기량과 실린더 개수 비례관게?

Mpg와 disp (-0.84)

Wt, disp(0.88): 엔진 배기량과 무게 비례관계?

wt와 cyl (0.78) 무게와 실린더 개수 비례


cyl,hp(0.83): 실린더 개수 비슷하면 엔진 최대 출력량 비슷하다?
Vs,cyl(-0.81): 엔진의 구조가 비슷하면 실린더 개수 역수 관계?

disp와 vs (-0.71)

```python
x = df['wt']
y = df['mpg']
plt.scatter(x, y)
plt.title("x:weight, y:mpg") 

import numpy as np

m, b = np.polyfit(x, y, 1)
#add linear regression line to scatterplot 
plt.plot(x, m*x+b)
#plt.show()
```
