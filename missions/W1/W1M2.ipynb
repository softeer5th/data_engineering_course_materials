```python
pd.read_sql("""SELECT Products.ProductID, Products.ProductName, Categories.CategoryName
                FROM Products
                INNER JOIN Categories ON Products.CategoryID = Categories.CategoryID;""",conn)
#JOIN 할 때는 Select할 때 컬럼명을 지정해 주는 것이 좋다. 
#특정 테이블에만 있는 컬럼이면 상관 없는데, 두 테이블에 공통으로 있는 컬럼을 조회할 때 테이블명 언급 안 해주면 오류 난다.
#JOIN 의 디폴트는 INNER JOIN이다.
```

```python
#JOIN Three Tables
pd.read_sql("""SELECT Orders.OrderID, Customers.CustomerName, Shippers.ShipperName
                FROM (Orders
                INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID)
                INNER JOIN Shippers ON Orders.ShipperID = Shippers.ShipperID;""",conn)
```
