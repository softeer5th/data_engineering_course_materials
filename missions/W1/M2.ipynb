```python
import sqlite3

conn = sqlite3.connect('./data.db')

cursor = conn.cursor()

sql = open('./create.sql', 'r').read()

def init_db():
    cursor.executescript(sql)

def execute_and_print_query(query, limit=5):
    cursor.execute(query)
    print(query)
    i = 0
    for row in cursor:
        if i >= limit:
            print('...')
            break
        print(row)
        i += 1
    print('-' * 50)

init_db()
```

```python
# SQL Syntax
execute_and_print_query('SELECT * FROM Customers')
```

```python
# SQL Union
execute_and_print_query("""SELECT City FROM Customers
UNION
SELECT City FROM Suppliers
ORDER BY City
""")
execute_and_print_query("""SELECT City FROM Customers
UNION ALL
SELECT City FROM Suppliers
ORDER BY City
""")
execute_and_print_query("""SELECT City, Country FROM Customers
WHERE Country='Germany'
UNION
SELECT City, Country FROM Suppliers
WHERE Country='Germany'
ORDER BY City
""")
execute_and_print_query("""SELECT City, Country FROM Customers
WHERE Country='Germany'
UNION ALL
SELECT City, Country FROM Suppliers
WHERE Country='Germany'
ORDER BY City""")
execute_and_print_query("""SELECT 'Customer' AS Type, ContactName, City, Country
FROM Customers
UNION
SELECT 'Supplier', ContactName, City, Country
FROM Suppliers""")
```
