# Sqlite Tutorial 실습

```python
import sqlite3
try:
    with sqlite3.connect("Northwind.db") as conn:
        print(f"Opened SQLite database with version {sqlite3.sqlite_version} successfully.")

except sqlite3.OperationalError as e:
    print("Failed to open database:", e)
```

```python
cursor = conn.cursor() #커서 생성
```

```python
def runSQL(SQL: str):
    cursor.execute(SQL)
    return cursor.fetchall()
```

LEFT JOIN -> 왼쪽 테이블의 데이터를 반환 (양측 공통 포함)

# FROM Customers 인데 Orders.OrderID가 어떻게 가능한거지?

LEFT JOIN을 사용하면, 왼쪽 테이블(이 경우 Customers)의 각 행에 대해 오른쪽 테이블(이 경우 Orders)의 일치하는 행을 결합합니다. <br>
그러므로 Customers 테이블의 컬럼 외에도, **Orders 테이블의 컬럼들이 함께 결과에 포함됩니다.**

```python
runSQL(
    '''
    SELECT Customers.CustomerName, Orders.OrderID
    FROM Customers
    LEFT JOIN Orders ON Customers.CustomerID = Orders.CustomerID
    ORDER BY Customers.CustomerName;
    '''
)
```

RIGHT JOIN -> LEFT JOIN과 반대

```python
runSQL(
    '''
    SELECT Orders.OrderID, Employees.LastName, Employees.FirstName
    FROM Orders
    RIGHT JOIN Employees ON Orders.EmployeeID = Employees.EmployeeID
    ORDER BY Orders.OrderID;
    '''
)
```

FULL OUTER JOIN -> FULL JOIN과 동일, 왼쪽 혹은 오른쪽에서 일치하는 레코드를 모두 반환

# 왜 CustomerName = Null인 행은 출력이 안되는거지???

예제 사이트와 내 학습용 데이터가 약간 차이가 있어 결과물에 차이가 있는 것으로 결론이 났다


```python
runSQL(
    '''
    SELECT Customers.CustomerName, Orders.OrderID
    FROM Customers
    FULL OUTER JOIN Orders ON Customers.CustomerID=Orders.CustomerID
    ORDER BY Customers.CustomerName;
    '''
)
```

Self Join -> 스스로에게 조인을 건다 같은 테이블이므로 Alias를 다르게 하여 구분

# <> 이건 뭐지? A.City = B.City --> 그냥 같은거 아닌가..?
--> <> 는 != 를 의미함

```python
runSQL(
    '''
    SELECT A.CustomerName AS CustomerName1, B.CustomerName AS CustomerName2, A.City
    FROM Customers A, Customers B
    WHERE A.CustomerID <> B.CustomerID
    AND A.City = B.City -- 한 도시에 사는 다른 사람?을 뽑는 듯
    ORDER BY  A.City;
    '''
)
```

ANY, ALL<br>
ANY --> 불 값 리턴, 서브쿼리 값이 하나라도 조건에 맞으면 참 <br>
ALL --> 불 값 리턴, 서브쿼리 값이 전부 조건에 맞으면 참, SELECT, WHERE, HAVING과 쓰임

```python
# ANY --> 안된다! SQLite는 ANY를 지원하지 않기 때문에 IN으로 대체해야 한다.
runSQL(
    '''
    SELECT ProductName 
    FROM Products
    WHERE ProductID = ANY (SELECT ProductID FROM OrderDetails WHERE Quantity = 10);
    '''
)
```

이렇게 하면 돌아간다

```python
runSQL(
    '''
    SELECT ProductName 
    FROM Products
    WHERE ProductID IN (SELECT ProductID FROM OrderDetails WHERE Quantity = 10);
    '''
)
```

```python
# ALL --> SQLite에서 지원 안함
runSQL(
    '''
    SELECT ProductName
    FROM Products
    WHERE ProductID = ALL (
        SELECT ProductID
        FROM OrderDetails
        WHERE Quantity = 10
    );
    '''
)
```

대체 방법 1. MAX(), MIN() 사용하기 <br>
MAX() 함수 사용하기 ALL은 서브쿼리의 모든 값과 비교를 요구하므로, 서브쿼리의 최대값을 구해 조건으로 사용

대체 방법 2. NOT EXISTS 사용하기 -> 모든 값에 대해 조건을 만족해야 하므로 이를 NOT EXISTS로 표현할 수 있다

```python
runSQL(
    '''
    SELECT ProductName
    FROM Products p
    WHERE NOT EXISTS (
        SELECT 1
        FROM OrderDetails od
        WHERE od.Quantity = 10 AND od.ProductID <> p.ProductID);
    '''
)
```

CASE --> if문 같은 역할, 조건 통과 실패 시 ELSE의 결과를 뱉고 ELSE가 없다면 NULL return

```python
runSQL(
    '''
    SELECT OrderID, Quantity,
    CASE
        WHEN Quantity > 30 THEN 'The quantity is greater than 30'
        WHEN Quantity = 30 THEN 'The quantity is 30'
        ELSE 'The quantity is under 30'
    END AS QuantityText
    FROM OrderDetails;
    '''
)
```

```python
runSQL(
    '''
    SELECT CustomerName, City, Country
    FROM Customers
    ORDER BY
    (CASE
        WHEN City IS NULL THEN Country
        ELSE City
    END);    

    '''
)
```
