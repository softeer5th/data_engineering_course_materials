### crosstab 사용

```python
# 1. crosstab을 이용하여 gear와 am에 대한 테이블을 생성
df_crosstable = pd.crosstab(df['gear'], df['am'])
print(df_crosstable)

# 개선 방법 찾기. 이게 최선인가?
comb = {}
for row in df_crosstable.iterrows():
    for idx, value in enumerate(row[1]):
        comb[f"({row[0]}, {idx})"] = value

plt.title("# of Cars by Gear x Transmission")
plt.xlabel("(Gear, Auto Transmission)")
plt.ylabel("# of Cars")
plt.bar(comb.keys(), comb.values())
plt.show()
```

### crosstab, 그러나 stack을 통한 더 나은 풀이

```python
# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# 바 그래프 생성
df_crosstable.stack().plot(kind='bar')

# 그래프 제목과 레이블 설정
plt.title('# of Cars by Gear x Transmission')
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')

# x축 레이블 회전 (보기 편하게)
plt.xticks(rotation=45)

# 그래프 여백 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

### crosstab 미사용

```python
gear2am = df.groupby(['gear', 'am']).size().unstack().fillna(0)
# groupby를 이용하여 gear와 am에 대한 테이블을 생성, size()로 각 그룹별 데이터 계산 후, unstack()으로 테이블 형태로 변환.
# NaN값을 0으로 채움.
# size()까지만 하면 Series로 반환됨.

print(gear2am)

comb = {}
for row in df_crosstable.iterrows():
    for idx, value in enumerate(row[1]):
        comb[f"({row[0]}, {idx})"] = value

# 그래프 그리기
plt.title("# of Cars by Gear x Transmission")
plt.xlabel("(Gear, Auto Transmission)")
plt.ylabel("# of Cars")
plt.bar(comb.keys(), comb.values())
plt.show()
```

## Histograms of Variables

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# 그래프 설정
column_counts = len(df.columns)  # column 수: 12
nrows = 4
ncols = 3

fig = plt.figure(figsize=(10, 6))
fig.suptitle('Histograms of Variables', fontsize=16)
gs = gridspec.GridSpec(nrows, ncols, figure=fig)

df_no_model = df.drop(columns=['model'])
for idx, col in enumerate(df_no_model.columns):
    if idx == column_counts - 1:
        break

    ax = fig.add_subplot(gs[idx // ncols, idx % ncols])
    ax.hist(df_no_model[col])
    ax.set_title(col)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

## 상관계수 분석

```python
df_no_model_corr = df_no_model.corr('pearson')
df_no_model_corr
```

### 엄밀히 말하면 연속형과 연속형이 아닌 변수는 상관관계 분석을 위하여 다른 분석을 진행해야 한다.

## 상관관계 순위 우선 출력

```python
# 상관계수 행렬의 상삼각 부분만 리스트 컴프리헨션으로 추출
pairs = [(i, j, df_no_model_corr.loc[i, j]) 
         for i in df_no_model_corr.index 
         for j in df_no_model_corr.columns if i < j]
# print(pairs)
# 정렬 후 상관관계가 높은 변수 쌍 n개까지 출력
n = 15
print(f"[Top {n} pairs with highest correlation]")
sorted_pairs = sorted(pairs, key=lambda x: abs(x[2]), reverse=True)

for i, pair in enumerate(sorted_pairs[:n], 1):
    print(f"{i}th: ({pair[0]}, {pair[1]}, {pair[2]:.6f})")
```

## 팀 활동 요구사항

### 이런 데이터셋을 분석해서 얻을 수 있는 경제적 가치는 무엇일까요? 어떤 비즈니스 상황에서 이런 데이터셋을 분석해서 어떤 경제적 가치를 얻을 수 있을까요?

Q) 이런 데이터셋을 분석해서 얻을 수 있는 경제적 가치는 무엇일까요?

A) mtcars.csv 데이터셋에는 차량의 기능적인 스펙만 존재하며, 가격 및 사용자 경험과 같은 지표가 포함되어 있지 않습니다. 그렇기 때문에 데이터 분석을 통해 연비나 속도 같은 기능적 요소 중에서 자사가 타사 대비 비교 우위인 속성이 어떤 것이 있는지 찾는 것이 경제적 가치를 얻을 수 있는 방법이라고 생각하였습니다. 또한 데이터셋을 통해 변수들 간의 상관계수를 구하기 전에, 기본적으로 차량의 용도와 종류에 대한 기본적인 분류가 필요하다는 생각도 들었습니다. 가령 차량들을 소형, 준중형, 중형, 대형 등으로 나누고, 만들고자 하는 차량에 대해 더욱 집중적으로 분석을 하거나, 각각에 대해 분석을 진행한 이후 더욱 경쟁력 있는 제품을 만들어 낼 수 있는 차종을 선택하는 것이 합리적일 것일 것입니다. 이후 기능별 예상 제작비용 등의 데이터셋이 추가된다면 더욱 정밀한 데이터 분석이 가능하고, 경제적 가치를 극대화할 수 있을 것입니다.

Q) 어떤 비즈니스 상황에서 이런 데이터셋을 분석해서 어떤 경제적 가치를 얻을 수 있을까요?

A) 사용자 경험과 비용에 대한 자료가 존재하지 때문에, 신차 제작 극초기 단계 기획에서 차종 별로 타사 대비 자사의 비교 우위가 어떤 것이 있는지 파악하면서 제일 성공률이 높은 차종 선택과 정량적 스펙을 설정할 수 있습니다. 따라서, 초기에 정확한 시장 세그먼트를 설정할 때 경제적 가치를 가질 수 있습니다. 특히, 데이터셋에 존재하는 속성 중 타사 대비 장점이라고 확실하게 말할 수 있는 것은 속도이기에 qsec와 관련한 지표에 집중하여 제일 성공률이 높은 차종 선택과 스펙을 설정할 것 같습니다. (연비의 경우 무게와 높은 음의 상관관계를 보이지만, 무게는 차량의 용도에 따라 달라질 수 있고, 무게 외 다른 요소들과는 크게 높은 상관관계를 보이지 않아 배제하였습니다.)

### 변수들 간의 상관 관계가 높은 조합을 임의로 2개 선택해서 해당 데이터 간의 상관 관계를 그래프로 그리고 어떤 결론을 내릴 수 있는지를 토의하세요

1. 마력과 무게, 이에 따른 가속도 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt

source = 'https://raw.githubusercontent.com/softeerbootcamp4th/data_engineering_course_materials/refs/heads/main/missions/W1/mtcars.csv'
df = pd.read_csv(source)

# 마력 vs qsec 산점도
plt.figure(figsize=(6, 4))
plt.scatter(df['hp'], df['qsec'], alpha=0.7)
plt.title('Horsepower vs qsec')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Quarter mile time (qsec)')
plt.grid(True)
plt.show()

# 차량 중량 vs qsec 산점도
plt.figure(figsize=(6, 4))
plt.scatter(df['wt'], df['qsec'], alpha=0.7)
plt.title('Vehicle Weight vs qsec')
plt.xlabel('Vehicle Weight (wt)')
plt.ylabel('Quarter mile time (qsec)')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
scatter = plt.scatter(df['hp'], df['wt'], c=df['qsec'], cmap='viridis', alpha=0.8)
plt.colorbar(scatter, label='qsec')
plt.title('Horsepower and Weight vs qsec')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Vehicle Weight (wt)')
plt.grid(True)
plt.show()
```

- 결론적으로 차체에 무게가 높더라도, 마력이 높으면 가속도가 높아집니다. 
  
- 가속도가 높은 차는 차체의 무게가 있더라도 마력이 높은 상관관계를 보입니다.

2. 가속도와 V 형태 엔진 산점도 그래프

```python
fig, axes = plt.subplots(figsize=(6, 2))
df.plot(x='qsec', y='vs', kind='scatter', ax=axes)
axes.set_yticks([0, 1])

plt.suptitle("Quarter Mile Time vs V Shape Engine")
axes.set_ylim(-0.5, 1.5)
plt.show()
```

- 가속도가 높은 경우 대부분은 V shape인 경우(0이 V shape)입니다.
