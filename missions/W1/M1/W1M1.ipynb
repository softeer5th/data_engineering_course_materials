# gear x transmission 조합 구하기
##### gear와 transmission의 경우의 수 조합별로 나누고 각 조합에 몇개의 차량이 있는지를 알아내세요.
##### 먼저 결과값을 화면에 출력하세요.
##### 결과값을 바 그래프로 그리세요. 타이틀은 '# of Cars by Gear x Transmission'로, x축 레이블은 '(Gear, Auto Transmission)', y축 레이블은 '# of Cars'로 만드세요.
##### gear와 transmission의 경우의 수 조합별로 나누고 각 조합에 몇개의 차량이 있는지를 알아낼 때 
##### 1)pandas.crosstab method를 사용하는 방법
##### 2)사용하지 않는 방법
##### 2가지 모두의 결과값을 각각 출력해야 합니다.

```python
#crosstab 사용하는 방법

gear_transmission_counts = pd.crosstab(index=[df['gear'], df['am']], columns=df['am'])
print("Number of cars by Gear x Transmission:")
print(gear_transmission_counts)
```

```python
gear_transmission_counts.plot(kind='bar')
```

```python
# 사용하지 않는 방법 -> groupby
gear_transmission_counts = df.groupby(['gear', 'am']).size()

print("Number of cars by Gear x Transmission:")
print(gear_transmission_counts)
```

```python
gear_transmission_counts.plot(kind='bar', title='# of Cars by Gear x Transmission', ylabel='# of Cars',
         xlabel='(Gear, Auto Transmission)', figsize=(6, 5), color='skyblue')
```

##### 데이터셋에 포함된 모든 features의 히스토그램을 그리세요.
##### 그래프 타이틀은 "Histograms of Variables"로 하세요.
##### 최대한 잘 보이도록 그래프들을 배치하세요.

```python
import matplotlib.pyplot as plt
columns = df.columns #[1:]

# 서브플롯 배치 설정
num_cols = 3  # 한 행에 들어갈 그래프 수
num_rows = (len(columns) + num_cols - 1) // num_cols  # 행의 개수 계산

# 그래프 크기 설정
plt.figure(figsize=(15, num_rows * 4))
plt.suptitle("Histograms of Variables", fontsize=16)

# 각 수치형 변수에 대해 히스토그램 생성
for i, col in enumerate(columns, 1):
    plt.subplot(num_rows, num_cols, i)
    plt.hist(df[col], bins=10)
    plt.title(col, fontsize=12)
    plt.xlabel(col, fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.grid(True, alpha=0.5, linestyle='--')
    if col == 'name':
        plt.xticks(rotation=90)

# 레이아웃 최적화
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# 팀 활동 요구사항
- 이런 데이터셋을 분석해서 얻을 수 있는 경제적 가치는 무엇일까요? 어떤 비즈니스 상황에서 이런 데이터셋을 분석해서 어떤 경제적 가치를 얻을 수 있을까요?
- 변수들 간의 상관 관계가 높은 조합을 임의로 2개 선택해서 해당 데이터 간의 상관 관계를 그래프로 그리고 어떤 결론을 내릴 수 있는지를 토의하세요.

---
1. 시각화 결과나 상관계수를 보면서 요소들 간의 관계(비례, 반비례 등)을 알고 그걸 토대로 자동차를 제작하는 입장에서의 의사결정에 도움을 줄 수 있을 것 같다. 연비를 높이려면 마력을 포기한다던지 등. 다만 단순히 지금 연비를 높이면 마력이 낮아진다 정도를 예상할 수 있고, 연비와 마력을 동시에 높이고 싶다면 그 방법에 대해서는 다른 연구가 필요하다. 
   또 히스토그램이 높게 나타난 지점에 대해 소비자의 선호가 나타난 것으로 보고 어느 요소에 집중해야 잘 팔리는 차를 만들 수 있을지 알 수 있을 것이다.
3. 팀원의 아이디어에 따라 피처별 히스토그램을 구했을 때 너무 sparse한 데이터는 제외하고, 한 요소를 중점으로 타 요소와의 상관관계를 분석한다
   sparse 데이터를 제외한 것들 중 '연비'를 중점으로 배기량과 무게를 비교했다.


```python
#연비와 배기량 비교
#연비가 증가할수록 배기량은 줄어든다. 
#배기량이 높은 차량은 흡입할 수 있는 공기가 많기 때문에 엔진의 크기가 크고 소비되는 연료도 많음을 의미한다. 즉 소비 연료 적다 = 연비 높다 = 배기량 적다

plt.figure(figsize=(8,6))
plt.xlabel('Miles Per Gallon', fontsize = 10)
plt.ylabel('disp', fontsize = 10)
plt.title("MPG vs DISP", fontsize = 12)

plt.scatter(df['mpg'], df['disp'], color='blue', alpha=0.7)
```

```python
#연비와 무게 비교
#연비가 증가할수록 무게는 줄어든다
#당연히 무게가 가벼운 차 일수록 연료가 적게 소모될 것이다.

plt.figure(figsize=(8,6))
plt.xlabel('Miles Per Gallon', fontsize = 10)
plt.ylabel('Weight', fontsize = 10)
plt.title("MPG vs Weight", fontsize = 12)

plt.scatter(df['mpg'], df['wt'], color='blue', alpha=0.7)
```
