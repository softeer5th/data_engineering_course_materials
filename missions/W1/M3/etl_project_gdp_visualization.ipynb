```python
import pandas as pd

# ETL processed data path
DATA_PATH = "data/Countries_by_GDP_etl_processed.json"
```

```python
# Load the data
with open(DATA_PATH, "r") as f:
    df = pd.read_json(f, dtype={"year": "Int32"})

display(df)
```

```python
# DataFrame of countries with GDP of 100(Billion) or more

df_100 = df[df["gdp"] >= 100]

display(df_100)
```

```python
# Mean GDP of top 5 countries by GDP in each region (sorted in descending order)

df_top5_mean = (
    df.groupby("region")["gdp"]
    .apply(lambda x: x.nlargest(5).mean())
    .sort_values(ascending=False)
)

display(df_top5_mean)
```
